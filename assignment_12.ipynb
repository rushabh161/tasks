```python
#Q2
def analyse_string(s):

    if s =="":
        return "The string is empty."
    
    print("Length of the string:", len(s))
    print("Reversed string:", s[::-1])


    vowels = "aeiouAEIOU"
    count = 0
    for ch in s:
        if ch in vowels:
            count += 1
            print("Vowel count in s:", count)

            for i in range(len(s)):
                print(f"Char: {s[i]}, Positive {i}, Negative {i - len(s)}")
            

s = input("Enter a string: ")
analyse_string(s)
```
