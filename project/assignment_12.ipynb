```python
#Q2
def   analyze_string ( s ) :       
    if  s ==  "" :           
        print ( "Empty string entered" )           
        return       
    print ( "Length:" ,   len ( s ))       
    print ( "Reverse:" ,  s [:: -1 ])      
    
    vowels =  "aeiouAEIOU"      
    count =  0       
    
    for  ch  in  s :           
        if  ch  in  vowels :              
            count +=  1       
            print ( "Vowels count:" ,  count )       
            for  i  in   range ( len ( s )):
                print(f"Char:  { s [ i ]} , Positive:  { i } , Negative:  { i -  len ( s )} ")  
                
                
s =  input ( "Enter a string: " )  
analyze_string ( s )  
```
